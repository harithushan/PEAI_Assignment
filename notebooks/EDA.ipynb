```python
# 4. Rajarata Release Analysis - Focus on the main topic
# Calculate 3-month rolling average to smooth daily variations
df['RAJARATA_3M_AVG'] = df['RAJARATA_POWER_RELEASE(MCM)'].rolling(window=90).mean()

plt.figure(figsize=(15, 12))

# Rajarata release over time with trend
plt.subplot(3, 1, 1)
plt.plot(df.index, df['RAJARATA_POWER_RELEASE(MCM)'], color='green', alpha=0.5, label='Daily')
plt.plot(df.index, df['RAJARATA_3M_AVG'], color='darkgreen', linewidth=2, label='3-month Avg')
plt.title('Rajarata Water Release (2000-2022)')
plt.ylabel('Release (MCM)')
plt.legend()
plt.grid(True, alpha=0.3)
plt
```

```python
# Yearly total Rajarata release
# Set 'DATE' as the index before resampling
df = df.set_index('DATE')

yearly_rajarata = df.resample('Y')['RAJARATA_POWER_RELEASE(MCM)'].sum()
plt.subplot(3, 1, 2)
plt.bar(yearly_rajarata.index.year, yearly_rajarata, color='green')
plt.title('Yearly Total Rajarata Release')
plt.xlabel('Year')
plt.ylabel('Total Release (MCM)')
plt.grid(True, alpha=0.3)
plt
```

```python
# Rajarata release as percentage of total
yearly_pct = (df.resample('Y')['RAJARATA_POWER_RELEASE(MCM)'].sum() /
              df.resample('Y')['TOTAL_RELESE(MCM)'].sum() * 100)
plt.subplot(3, 1, 3)
plt.plot(yearly_pct.index.year, yearly_pct, 'o-', color='green')
plt.axhline(y=yearly_pct.mean(), color='red', linestyle='--', label=f'Average: {yearly_pct.mean():.1f}%')
plt.title('Rajarata Release as Percentage of Total Release')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.ylim(0, max(yearly_pct) * 1.1)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('rajarata_release_analysis.png', dpi=300)
# plt.close()
plt
```

```python






# 6. Water Balance Analysis - Inflow vs Outflow and Storage
df['STORAGE_CHANGE'] = df['STORAGE_(MCM)'].diff()
df['EXPECTED_CHANGE'] = df['INFLOW(MCM)'] - df['TOTAL_RELESE(MCM)']
df['BALANCE_DIFFERENCE'] = df['STORAGE_CHANGE'] - df['EXPECTED_CHANGE']

# Yearly water balance
yearly_balance = df.resample('Y').sum()
yearly_balance['INFLOW_TOTAL'] = yearly_balance['INFLOW(MCM)']
yearly_balance['OUTFLOW_TOTAL'] = yearly_balance['TOTAL_RELESE(MCM)']
yearly_balance['BALANCE'] = yearly_balance['INFLOW_TOTAL'] - yearly_balance['OUTFLOW_TOTAL']

plt.figure(figsize=(15, 10))

# Yearly inflow vs outflow
plt.subplot(2, 1, 1)
width = 0.35
x = np.arange(len(yearly_balance.index))
plt.bar(x - width/2, yearly_balance['INFLOW_TOTAL'], width, label='Inflow')
plt.bar(x + width/2, yearly_balance['OUTFLOW_TOTAL'], width, label='Outflow')
plt.title('Yearly Water Inflow vs Outflow')
plt.xlabel('Year')
plt.ylabel('Volume (MCM)')
plt.xticks(x, [d.year for d in yearly_balance.index], rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt
# inflow vs *rajarata
```

```python
# Yearly water balance
plt.subplot(2, 1, 2)
plt.bar(yearly_balance.index.year, yearly_balance['BALANCE'], color='blue' if yearly_balance['BALANCE'].mean() > 0 else 'red')
plt.axhline(y=0, color='black', linestyle='-')
plt.title('Yearly Water Balance (Inflow - Outflow)')
plt.xlabel('Year')
plt.ylabel('Balance (MCM)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('yearly_water_balance.png', dpi=300)
# plt.close()

plt
```

```python
# 7. Focus on Farming Seasons - Rajarata Release during key agricultural periods
# Define typical farming seasons in Sri Lanka (approximate)
# Maha season: October to March (wet season)
# Yala season: April to September (dry season)

df['SEASON'] = 'Maha (Wet)'  # October to March
df.loc[(df.index.month >= 4) & (df.index.month <= 9), 'SEASON'] = 'Yala (Dry)'

# Seasonal analysis
season_group = df.groupby(['YEAR', 'SEASON']).sum().reset_index()

# Pivot the data for easier plotting
season_pivot = season_group.pivot(index='YEAR', columns='SEASON',
                                 values='RAJARATA_POWER_RELEASE(MCM)')

plt.figure(figsize=(15, 10))

# Rajarata release by season and year
plt.subplot(2, 1, 1)
plt.plot(season_pivot.index, season_pivot['Maha (Wet)'], 'o-', color='blue', label='Maha (Wet) Season')
plt.plot(season_pivot.index, season_pivot['Yala (Dry)'], 'o-', color='orange', label='Yala (Dry) Season')
plt.title('Rajarata Water Release by Farming Season')
plt.xlabel('Year')
plt.ylabel('Total Release (MCM)')
plt.legend()
plt.grid(True, alpha=0.3)
plt
```

```python
# Add season column
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['SEASON'] = df['MONTH'].apply(get_season)
```

```python
# Time Series Analysis
print("\nTime Series Analysis:")

# Monthly averages
monthly_avg = df.groupby('MONTH').agg({
    'WATER_LEVEL(MSL)': 'mean',
    'STORAGE_(MCM)': 'mean',
    'STORAGE__PERCENTAGE': 'mean',
    'RAINFALL(mm)': 'mean',
    'ENERGY(MWh)': 'mean',
    'INFLOW(MCM)': 'mean',
    'TOTAL_RELESE(MCM)': 'mean'
}).reset_index()

print("\nMonthly Averages:")
print(monthly_avg)
```

```python
# Yearly averages
yearly_avg = df.groupby('YEAR').agg({
    'WATER_LEVEL(MSL)': 'mean',
    'STORAGE_(MCM)': 'mean',
    'STORAGE__PERCENTAGE': 'mean',
    'RAINFALL(mm)': 'sum',
    'ENERGY(MWh)': 'sum',
    'INFLOW(MCM)': 'sum',
    'TOTAL_RELESE(MCM)': 'sum'
}).reset_index()

print("\nYearly Aggregates:")
print(yearly_avg)
```

```python
# Seasonal analysis
seasonal_avg = df.groupby('SEASON').agg({
    'WATER_LEVEL(MSL)': 'mean',
    'STORAGE_(MCM)': 'mean',
    'STORAGE__PERCENTAGE': 'mean',
    'RAINFALL(mm)': 'mean',
    'ENERGY(MWh)': 'mean',
    'INFLOW(MCM)': 'mean',
    'TOTAL_RELESE(MCM)': 'mean'
}).reset_index()

print("\nSeasonal Averages:")
print(seasonal_avg)
```

```python
# Calculate water balance efficiency
df['WATER_BALANCE_EFFICIENCY'] = (df['INFLOW(MCM)'] - df['TOTAL_RELESE(MCM)']) / df['INFLOW(MCM)'] * 100
print("\nWater Balance Efficiency Statistics:")
print(df['WATER_BALANCE_EFFICIENCY'].describe())
```

```python
# Calculate energy efficiency (energy per unit of water released)
df['ENERGY_EFFICIENCY'] = df['ENERGY(MWh)'] / df['TOTAL_RELESE(MCM)']
print("\nEnergy Efficiency Statistics (MWh per MCM):")
print(df['ENERGY_EFFICIENCY'].describe())
```
